```python
import pandas as pd
import matplotlib.pyplot as plt
import time
"""
La funcion medidor_tiempo se encarga de medir el tiempo de ejecucion de un programa.
"""
def medidor_tiempo(funcion):
    def funcion_medida(*args, **kwargs):
        inicio = time.time()*10**6
        c = funcion(*args, **kwargs)
        fin = time.time()*10**6
        #Se resta el segundo que puse en el sleep
        print("\nLa funcion: ",funcion.__name__," , se ejecuto en: {} micro segundos".format((fin - inicio)*(10**6)))
        return c
    return funcion_medida
```

### **1. Merge two sorted lists.**

#### &ensp; 1. Suponga que tiene dos listas enlazadas ordenadas. Utilice la recursividad para fusionar estas listas y devolver una nueva lista ordenada.
Por ejemplo:
- Input:
<br>L1 = 1 -> 3 -> 5 -> 7 -> nulo
<br>L2 = 2 -> 4 -> 6 -> nulo
- Output:
<br>L4 = 1 -> 2 -> 3-> 4 -> 5 -> 6 -> 7 -> nulo

##### &ensp; Literal 1

```python
"""
Clase que crea los nodos de la linked list
"""
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
"""
clase que crea la singly linked list
"""
class Singly_linked_list:
    """
    Funcion que inicializa la cabeza de la lista enlazada -linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
        
    """
    Funcion que recorre la lista enlazada - linked list
    """
    @medidor_tiempo
    def list_traversed(self):
        time.sleep(1)
        node = self.head_node
        while node:
            print(node.val,"->" , end = " ")
            node = node.next_node 
        print("Null")
        
    """
    Funcion que inserta al final de la linked list
    """
    @medidor_tiempo
    def insert_tail(self, new_node):
        time.sleep(1)
        node = self.head_node
        prev = None
        while node:
            prev = node
            node = node.next_node
        prev.set_next_node(new_node) 

"""
Funcion que ordena la lista de numeros siempre y cuando las listas que se le
entreguen esten ordenadas
"""
@medidor_tiempo
def merge_sort(list1,list2):
    time.sleep(1)
    if list1 is None:
        return list2
    elif list2 is None:
        return list1
     
    if list1.val < list2.val:
        list1.next_node = merge_sort(list1.next_node, list2)
        return list1
    else:
        list2.next_node = merge_sort(list1, list2.next_node)
        return list2
```

```python
#Se crea la linked list 1
list1 = Singly_linked_list(Node(2))
list1.insert_tail(Node(6))
list1.insert_tail(Node(8))
print ("List 1: ")
list1.list_traversed()
```

```python
#Se crea la linked list 2
list2 = Singly_linked_list(Node(1))
list2.insert_tail(Node(3))
list2.insert_tail(Node(4))
list2.insert_tail(Node(5))
list2.insert_tail(Node(7))
print ("\n\nList 2: ")
list2.list_traversed()
```

```python
#Se crea la linked list 3 que contenga el contenido de list 1 y list 2
list3 = Singly_linked_list(merge_sort(list1.head_node, list2.head_node))
print ("\n\nList 3: ")
list3.list_traversed()
```

### **2. Implement quick sort.**

#### &ensp; (Implementar clasificación rápida)
#### &ensp; 1. Usa la técnica divide y vencerás para implementar el algoritmo de clasificación rápida estudiado en clase.
#### &ensp; 2. Use un par de párrafos para explicar los pasos de divide y vencerás:
- ¿cómo/dónde se divide?
- ¿dónde/qué conquistas (conquer)?
- ¿cómo/qué combinas (combine)? 

#### &ensp; 3. Analizar la complejidad

##### &ensp; Literal 1

```python
@medidor_tiempo
def quick_sort(array):
    time.sleep(1)
    #Si el arreglo tiene longitud 1, ya esta ordenad
    if len(array) <=1:
        arreglo_ordenado = array
    else:
        #Se selecciona el pivote que sera el primer elemento del arreglo
        pivote = array[0]
        #Se inicializan 3 arreglos para elementos mayores, menores e iguales
        greater = []
        less = []
        equal = [] 
        for i in array:
            if i > pivote:
                greater.append(i)
            elif i < pivote:
                less.append(i)
            else: 
                equal.append(i)
        arreglo_ordenado = quick_sort(less)+equal+quick_sort(greater)
    return arreglo_ordenado
```

```python
quick_sort([2, 1, 3, 5, 6, 3, 8, 10])
```

##### &ensp; Literal 2

**Divide and Conquer:**
<br> &ensp;**1.** **Divide:** Se elige un pivote. Los elementos menores que el pivote se ubicarán en el lado izquierdo y los elementos mayores al pivote estarán en el lado derecho del pivote.
<br> &ensp;**2.** **Conquer:** Se manejará un proceso recursivo, para ordenar el arreglo.
<br> &ensp;**3.** **Combine:** Al ya estar ordenados los elementos se combinan y forman un arreglo ordenado.

##### &ensp; Literal 3

**Análisis:**
<br> &ensp;**1.** La funcion `quick_sorted` tiene con una complejidad de tiempo $O(n^2)$: complejidad cuadrática y una complejidad de espacio $O(log(n))$:complejidad logarítmica.

### **3. Compare quick sort and merge sort**
#### &ensp;  (Comparar quick sort y merge sort)
#### &ensp; 1. Genere 1000 permutaciones aleatorias de matrices de enteros entre 1 y 10000.
#### &ensp; 2. Ejecute merge sort y quick sort para cada una de las permutaciones
#### &ensp; 3. Calcule el tiempo de ejecución de estas ejecuciones y compárelas utilizando las estadísticas del tiempo de ejecución medido.
#### &ensp; 4. Explique las diferencias. ¿Qué algoritmo es mejor? ¿Cuándo usarías cada uno?

##### &ensp; Literal 1

```python
import numpy as np
arrays = []
for i in range(0,1000):
    array = np.random.permutation(10000)
    arrays.append(array)
    print(arrays[i])
```

##### &ensp; Literal 2 Y 3

```python
"""
La funcion contador_tiempo se encarga de medir el tiempo de ejecucion de un programa.
"""
def contador_tiempo(funcion):
    def funcion_medida(*args, **kwargs):
        fin = time.time()*10**6
        return abs((fin-time.time()))
    return funcion_medida
```

```python
@contador_tiempo
def merge_sort(nums):
    time.sleep(1)
    if len(nums) <= 1:
        return nums
    pivot = int(len(nums) / 2)
    left = merge_sort(nums[0:pivot])
    right = merge_sort(nums[pivot:])
    return merge(left, right)


def merge(left, right):
    left_pointer = 0
    right_pointer = 0
    sorted_list = []
    while left_pointer < len(left) and right_pointer < len(right):
        if left[left_pointer] < right[right_pointer]:
            sorted_list.append(left[left_pointer])
            left_pointer += 1
        else:
            sorted_list.append(right[right_pointer])
            right_pointer += 1
    
    sorted_list.extend(left[left_pointer:])
    sorted_list.extend(right[right_pointer:])
    
    return sorted_list
```

```python
timeMS= []
for i in range(0,1000):
    timeMS.append(merge_sort(arrays[i]))
    print(timeMS[i])
```

```python
@contador_tiempo
def quick_sort(array):
    time.sleep(1)
    #Si el arreglo tiene longitud 1, ya esta ordenad
    if len(array) <=1:
        arreglo_ordenado = array
    else:
        #Se selecciona el pivote que sera el primer elemento del arreglo
        pivote = array[0]
        #Se inicializan 3 arreglos para elementos mayores, menores e iguales
        greater = []
        less = []
        equal = [] 
        for i in array:
            if i > pivote:
                greater.append(i)
            elif i < pivote:
                less.append(i)
            else: 
                equal.append(i)
        arreglo_ordenado = quick_sort(less)+equal+quick_sort(greater)
    return arreglo_ordenado
```

```python
timeQS= []
for i in range(0,1000):
    timeQS.append(quick_sort(arrays[i]))
    print(timeMS[i])
```

```python
df = pd.DataFrame(zip(timeMS,timeQS))
df.columns = [ 'Merge Sort','Quick Sort']
df
```

```python
plt.plot(df)
plt.yscale('log')
plt.show()
```

##### &ensp; Literal 4

**Análisis:**
<br> &ensp;**Semejanzas:**
<br> &ensp;&ensp;**1.**`Merge Sort` y `Quick Sort` son métodos de ordenamiendo.
<br> &ensp;&ensp;**2.**`Merge Sort` y `Quick Sort` usan el principio de divide and conquer.
<br> &ensp;&ensp;**3.**`Merge Sort` y `Quick Sort` usan recursividad,
<br> &ensp;**Diferencias:**
<br> &ensp;&ensp;**1.** La complejidad de `Merge sort` es de $O(nlogn)$ mientras que la complejidad de `Quick sort` es de $O(n^2)$. 
<br> &ensp;&ensp;**2.** `Merge sort` divide el arreglo en 2 arreglos y esos 2 en otros dos y así sucesivamente, mientras que `Quick sort` tiene un pivote con el que se compara cada elemento, los elementos con un valor menor que el pivote van al lado izquierdo del pivote, mientras que los elementos con un valor mayor que el pivote van al lado derecho del pivote. 
<br> &ensp;&ensp;**3.** `Merge sort` funciona mejor con arreglos de tamaño más grande a diferencia de `Quick Sort` que maneja de forma eficiente arreglos más pequeños.
<br> &ensp;**¿Qué algoritmo es mejor?:**
<br> &ensp;&ensp; Por la complejidad del algoritmo se conluye que `Merge Sort`es un algoritmo más eficiente.
<br> &ensp;**¿Cuándo usarías cada uno?:**
<br> &ensp;&ensp; Como se mencionó anteriormente `Merge sort`sirve para arreglos de tamaño grande y `Quick Sort` para arreglos más pequeños.

### **4. Implement factorials with recursion**
####  &ensp; (Implementar factoriales con recursividad)
&ensp; Recuerda que un número factorial se define como:
$k! = k \times (k-1) \times (k-2) \times ... \times 2 \times 1$

&ensp; Por ejemplo:
&ensp; $3! = 3 \times 2 \times 1 = 6$

#### &ensp; 1. Usa la recursividad para implementar el factorial.
#### &ensp; 2. Ejecute su código para números entre 1 y 10000.
#### &ensp; 3. Calcular el tiempo de ejecución y trazarlo.
#### &ensp; 4. ¿Cuál es la complejidad?.
#### &ensp; 5. Implemente la memorización para mejorar el rendimiento de su código.
#### &ensp; 6. Mida el tiempo de ejecución cuando calcule el factorial para números entre 1 y 10000.
#### &ensp; 7. ¿Cuál es la complejidad de su código usando memoization?
#### &ensp; 8. ¿Cómo se comparan la primera implementación y la que tiene memorización?

##### &ensp; Literal 1

```python
@contador_tiempo
def factorial (num):
    time.sleep(1)
    if num == 0 or num == 1:
        resultado = 1
    elif num > 1:
        resultado = num*factorial(num-1)
    return resultado
```

##### &ensp; Literal 2

```python
factorial_times=[]
for i in range(1,10000):
    factorial_times.append(factorial(i))
```

##### &ensp; Literal 3

```python
df1 = pd.DataFrame(zip(factorial_times))
df1.columns = ['Factorial']
df1

df1.plot()
```

##### &ensp; Literal 4

**Análisis:**
<br> &ensp; &ensp; **1.** `factorial` con una complejidad de $O(n)$: complejidad lineal.

##### &ensp; Literal 5

```python
@contador_tiempo
def factorial1(num):
    cache=[]
    def factorial_Memoization(num):
        if num==0 or num == 1:
            res = 1
        if num in cache.keys():
            res = cache[num]
        else:
            res = num*factorial(num-1)
            cache[num] = res
        return res
    return factorial_Memoization(num)
```

##### &ensp; Literal 6

```python
factorialm_times=[]
for i in range(1,10000):
    factorialm_times.append(factorial1(i))
```

```python
df1 = pd.DataFrame(zip(factorialm_times))
df1.columns = ['Factorial Memoization']
df1

df1.plot()
```

##### &ensp; Literal 7

&ensp; &ensp; La complejidad para la función `factorial1` con memoización es $O(1)$: complejidad constante.

##### &ensp; Literal 8

&ensp; &ensp; La primera implementación tiene una complejidad lineal $O(n)$ y el algoritmo que tiene meoización  es de complejidad constante $O(1)$, lo que indical que el algoritmo que tiene memoización es más rápido dedido a que guarda cálculos anteriorespara calcular los siguientes.
